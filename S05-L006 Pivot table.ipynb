```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('data/marathon_results_2016.csv', usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'], index_col='Bib')
```

```python
df.head()
```

```python
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)
 
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
 
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
 
df['TotalSeconds'] = df['TotalSeconds'].replace(0,np.nan)
df['HalfSeconds'] = df['HalfSeconds'].replace(0,np.nan)
```

```python
df.head()
```

```python
df.pivot_table(columns='M/F', index='Age', values='TotalSeconds', aggfunc='mean').head()
```

```python
df.pivot_table(columns='M/F', index='Age', values=['TotalSeconds', 'HalfSeconds']).head()
```

```python
df.pivot_table(columns='M/F', index='Age', values=['TotalSeconds', 'HalfSeconds']).swaplevel(axis='columns').head()
```

```python
df.pivot_table(columns='M/F', index='Age', values=['TotalSeconds', 'HalfSeconds']).swaplevel(axis='columns').sort_index(axis='columns').head()
```
